# Test de segmentation d'une image

```python
from skimage.io import imread, imshow
from skimage.exposure import histogram
from scipy import ndimage as ndi
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 10
```

```python
img = imread('K389005_1144.png', True)
imshow(img)
print(img.shape)
```

## Masque

```python
img[img<80]=0
imshow(img)
```

## Morphologie mathmatique

```python
erode_img = ndi.binary_erosion(img)
reconstruction = ndi.binary_propagation(erode_img, mask=img)

plt.figure(figsize=(20, 10))
plt.subplot(131)
plt.imshow(img, cmap=plt.cm.gray, interpolation='nearest')
plt.subplot(132)
plt.imshow(erode_img, cmap=plt.cm.gray, interpolation='nearest')
plt.subplot(133)
plt.imshow(reconstruction, cmap=plt.cm.gray, interpolation='nearest')
plt.show()
```

## Rsultat

```python
imshow(reconstruction)
plt.contour(reconstruction, [0.5], linewidths=2, colors='r')
plt.show()
```
